testing with turtle - looks good:)

```python
import turtle

# Create a turtle object
t = turtle.Turtle()

# Create a screen object
screen = turtle.Screen()

actions = directions

# Iterate over the list of actions
for index, action in enumerate(actions):
    if action['action'] == 'Turn':
        if index == 0:
            t.left(action['value'] + 90)
        else:
            t.left(action['value'])
    elif action['action'] == 'Move':
        t.forward(action['value'])

# Exit the turtle graphics window
turtle.done()
```
